```python
import sqlite3

class EmpresaDB:
    def __init__(self, nome_banco):
        self.conn = sqlite3.connect(nome_banco)
        self.criar_tabela()

    def criar_tabela(self):
        self.conn.execute('''
        CREATE TABLE IF NOT EXISTS funcionarios (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            nome TEXT NOT NULL,
            cargo TEXT NOT NULL,
            salario REAL NOT NULL
        )
        ''')
        self.conn.commit()

    def inserir_funcionario(self, nome, cargo, salario):
        self.conn.execute('''
        INSERT INTO funcionarios (nome, cargo, salario) VALUES (?, ?, ?)
        ''', (nome, cargo, salario))
        self.conn.commit()

    def listar_funcionarios(self):
        cursor = self.conn.execute('SELECT * FROM funcionarios')
        for row in cursor:
            print(f'ID: {row[0]}, Nome: {row[1]}, Cargo: {row[2]}, Salário: {row[3]}')

    def buscar_funcionario_por_nome(self, nome):
        cursor = self.conn.execute('SELECT * FROM funcionarios WHERE nome = ?', (nome,))
        row = cursor.fetchone()
        if row:
            print(f'ID: {row[0]}, Nome: {row[1]}, Cargo: {row[2]}, Salário: {row[3]}')
        else:
            print('Funcionário não encontrado.')

    def fechar_conexao(self):
        self.conn.close()

# Exemplo de uso
empresa_db = EmpresaDB('empresa.db')

while True:
    print('\nOpções:')
    print('1. Inserir Funcionário')
    print('2. Listar Funcionários')
    print('3. Buscar Funcionário por Nome')
    print('4. Sair')

    opcao = input('Escolha uma opção: ')

    if opcao == '1':
        nome = input('Nome do Funcionário: ')
        cargo = input('Cargo do Funcionário: ')
        salario = float(input('Salário do Funcionário: '))
        empresa_db.inserir_funcionario(nome, cargo, salario)
        print('Funcionário inserido com sucesso!')
    elif opcao == '2':
        print('Lista de Funcionários:')
        empresa_db.listar_funcionarios()
    elif opcao == '3':
        nome = input('Digite o nome do Funcionário: ')
        empresa_db.buscar_funcionario_por_nome(nome)
    elif opcao == '4':
        empresa_db.fechar_conexao()
        break
    else:
        print('Opção inválida. Escolha novamente.')

print('Programa encerrado.')
```
